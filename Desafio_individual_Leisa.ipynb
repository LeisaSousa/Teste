```python
#4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)
```

```python
# Filtrar profissão advogado e representante de vendas
advogado_representante = df[df['Profissão'].isin(['Advogado', 'Representante de Vendas'])]

# Calcular a média de horas de sono para essas profissões
media_sono = advogado_representante ['Duração do sono'].mean()

# Comparar com a média geral
media_geral = df['Duração do sono'].mean()

print(f'Média de horas de sono para advogados ou representantes de vendas: {media_sono:.2f}')
print(f'Média geral de horas de sono: {media_geral:.2f}')
```

```python
#5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)
```

```python
# Filtrar profissão enfermagem e medicina
enfermagem_medicina = df[df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]

# Calcular a média de horas de sono para essas profissões
media_enfermagem_medicina = enfermagem_medicina.groupby('Profissão')['Duração do sono'].mean()

print("Média de horas de sono por profissão (Enfermagem x Medicina):")
print(media_enfermagem_medicina)
```
